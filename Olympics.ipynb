#### __8-Is the GDP per Capita effects on taking a medal or not ?__

```python
# Select the column that we need
GDP = olympics[['Country','GDP_per_Capita','Medal']]
```

```python
# this code works on taking a sample for the Athletes who have taken GDP above the median 50% of GDP
above_mean = GDP[GDP['GDP_per_Capita'] > GDP['GDP_per_Capita'].median()].sample(100)['Medal'].value_counts()
above_mean
```

```python
# this code works on taking a sample for the Athletes who have taken GDP below the median 50% of GDP
below_mean = GDP[GDP['GDP_per_Capita'] < GDP['GDP_per_Capita'].median()].sample(100)['Medal'].value_counts()
below_mean
```

```python
from plotly.subplots import make_subplots

above_mean = GDP[GDP['GDP_per_Capita'] > GDP['GDP_per_Capita'].median()].sample(frac=0.5)['Medal'].value_counts()
medal = above_mean.index # For change to list to helps in visualization
num = above_mean.unique() # For change to list to helps in visualization



below_mean = GDP[GDP['GDP_per_Capita'] < GDP['GDP_per_Capita'].median()].sample(frac=0.5)['Medal'].value_counts()
medal2 = below_mean.index # For change to list to helps in visualization
num2 = below_mean.unique() # For change to list to helps in visualization




fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]],subplot_titles=['Over mean GDP', "Below mean GDP"])

fig.add_trace(go.Pie(labels=medal, values=list, scalegroup='one' , name='Over'), 1, 1)

fig.add_trace(go.Pie(labels=medal2, values=num2,  scalegroup='one' , name='Below'), 1, 2 )

fig.update_traces(hoverinfo='label+percent+name', textinfo='label+percent', textfont_size=20 , hole = 0.5)

fig.update_layout(title_text = 'Random Sample to know GDP affects on take medale or not .',
annotations=[dict(text='Over %50 GDP', x=0.22, y=0.5, font_size=13, showarrow=False),
            dict(text='Below %50 GDP', x=0.77, y=0.5, font_size=13, showarrow=False)]) 


```

- Notes
  - Every time this code runs it provides change results.
  - I noticed that the Athletes have GDP below mean achieves medals greater than which have GDP over mean.
  - This code takes half the data whether above or below then take %50 as a random sample .
